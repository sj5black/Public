## 5.1 조건문

조건문은 특정 조건을 만족하는지 여부에 따라 실행하는 코드가 달라져야 할 때 사용.  
예를 들어 삼각형이 있을 시 삼각형 변의 길이에 따라  
정, 이등변, 직각, 일반 등의 삼각형으로 구분할 경우 조건이 필요.

### 5.1.1 if

파이썬은 이와 같은 상황에서 사용할 수 있는 if문 제공

사용법
* if 조건식:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식의 결과가 True일 때 실행문

<br>
조건식의 결과가 True 이면 실행문이 동작하고,
조건식의 결과가 False면 실행문이 동작하지 않음.
<br>

**※주의**
파이썬은 타 언어와 다르게 코드 블록을 나타내는  
중괄호({})가 없는 대신 들여쓰기로 구분.  
따라서 다음과 같이 쓸 경우 오류 발생.

* if 조건식:
<br>조건식의 결과가 True일 때 실행문


들여쓰기 규칙은 다음과 같음  
1. 공백이나 탭을 이용하여 들여쓰기 수행  
2. 공백의 개수는 상관 없음  
그러나 파이썬 스타일 가이드(PEP 8)에 따르면 공백의 개수는 4개 사용.  
3. 탭은 1개만 사용  
4. 동일 구역에서 들여쓰기는 통일  
(탭과 공백을 혼용하여 사용할 수 없고, 들여쓰기 수준도 동일)


```python
a = 3
if a > 3:
    print('a는 3보다 큰 수')

if a > 1:
    print('a는 1보다 큰 수')
```

### 5.1.2 if-else

if-else문은 조건식을 만족하는 경우와 만족하지 않는 경우를 
구분할 필요가 있을 시 사용.

사용법
* if 조건식:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식의 결과가 True일 때 실행문  
  else:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식의 결과가 False일 때 실행문

조건식의 결과가 True인 경우와 False인 경우로 구분하여 코드 작성 가능.

```python
a = 3

if a > 3:
    print('a는 3보다 큰 수')
else: # (a <= 3)
    print('a는 3보다 작거나 같은 수')
```

### 5.1.3 if-elif

if-elif문은 특정 조건에 따라 3가지 이상으로 구분해야 할 때 사용.
여기서 elif는 else if

사용법
* if 조건식1:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 True일 때 실행문  
  elif 조건식2:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 False이고 조건식2의 결과가 True일 때 실행문.  
  elif 조건식3:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2의 결과가 False이고 조건식 3의 결과가 True일 때 실행문

```python
a = 3

if a > 3:
    print('a는 3보다 큰 수')
elif a > 1: # (a<=3)
    print('a는 1보다 크고 3보다 작거나 같은 수')
```

### 5.1.4 if-elif-else

if-elif문은 특정 조건에 따라 구분 후 위의 조건을 만족하는 않는  
모든 경우에 대해 실행을 할 때 사용.

사용법
* if 조건식1:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 True일 때 실행문  
  elif 조건식2:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 False이고 조건식2의 결과가 True일 때 실행문.  
  elif 조건식3:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2의 결과가 False이고 조건식 3의 결과가 True일 때 실행문  
  else:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2, 3이 모두 False일 때 실행될 문장.

```python
a = 1

if a > 3:
    print('a는 3보다 큰 수')
elif a > 1: # (a<=3)
    print('a는 1보다 크고 3보다 작거나 같은 수')
else:
    print('a는 1보다 작거나 같은 수')
```

## 4. Grading

사용자로부터 점수를 입력받는다. <br>
입력받은 점수를 아래와 같이 분류하는 프로그램을 작성하시오. <br>

90 $\geq$ A <br>
80 $\geq$ B <br>
70 $\geq$ C <br>
60 $\geq$ D <br>
&nbsp;&nbsp;&nbsp;&nbsp;   < F <br>

```python
user_input = input()

if user_input.isdecimal():
    user_input = int(user_input)

    if 0 <= user_input <= 100:
        if user_input >= 90:
            print('A')
        elif user_input >= 80:
            print('B')
        elif user_input >= 70:
            print('C')
        elif user_input >= 60:
            print('D')
        else:
            print('F')
    else:
        print('입력한 숫자를 확인하세요.')
        print(f'현재 입력된 숫자는: {user_input}입니다.')
```

## 7. Codeup. 1229.

```python
height, weight = map(int, input().split())

if height < 150:
    standard_weight = height - 100
elif height < 160:
    standard_weight = (height-150)/2 + 50
else:
    standard_weight = (height-100) * 0.9

obesity = (weight-standard_weight)*100 / standard_weight

if obesity <= 10:
    print('정상')
elif obesity <= 20:
    print('과체중')
else:
    print('비만')
```
