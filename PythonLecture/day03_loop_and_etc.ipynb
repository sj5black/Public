## 5.2 반복문

반복문은 어떤 수행 작업을 한 번이 아니라 계속해서 수행해야 할 때 사용.  

반복문은 while과 for 두 가지 존재.

### 5.2.1 while

1.   while  
while문은 특정 조건을 만족하는 동안 반복해서 수행해야 하는 코드를 작성할 때 사용.  

* while문의 기본 구조는 다음와 같음  
while 조건식:  
     &nbsp;&nbsp;&nbsp;&nbsp;반복실행문
     
     조건식의 결과가 True이면 while문을 실행  
     반복문 실행이 종료되면 다시 조건식으로 이동  
     조건식의 결과가 False이면 더 이상 반복 실행문을 실행하지 않음
<br>

2.   while문이 필요한 경우  
while문은 주로 횟수가 정해지지 않은 연산을 진행할 때나  
어떤 특정 동작이 지속될 때 사용
<br>


3.   while문의 중첩
while문 내부에 또 다른 while문이 나타나는 것을 while문의 중첩이라고 함.  
반복처리해야 하는 대상이 2개 이상이면 while문의 중첩이 필요.  

     다음과 같은 예제서 사용  
     5일동안 매일 3시간씩 수업을 진행  
     매 시간 '1일차 1교시'와 같은 메세지를 출력  
     1일차부터 5일차까지 반복되는 일수와
     1교시부터 3교시까지 반복되는 시간이라는  
     2개의 반복처리 대상이 있으니 while문의 중첩 필요





<br>
<br>



```python
# total = 0
# for i in range(1, 101):
#     total += i

# -----------------
# while

total = 0
i = 1
while i < 101:
    total += i
    i += 1

print(total)
```

```python
total = 0
i = 1
while True:
    total += i

    if total > 100:
        break

print(total)
```

```python
# while문을 사용해서 hello world 10번 출력
# 아래는 출력 양식
# 1: hello world
# 2: hello world
# ...
# 10: hello world

i = 0
while i < 10:
    print(f'{i + 1}: hello world')
    i += 1
```

```python
# while문으로 구구단 9단의 1~9까지의 결과를 출력하는 코드 작성

dan = 9
i = 1
while i < 10:
    print(f'{dan} * {i} = {dan*i}')
    i += 1
```

```python
# 1 * 1 = 1  2 * 1 = 2      ...  9 * 1 = 9
# 1 * 2 = 2  2 * 2 = 4      ...  9 * 2 = 18
# .....      ...            ...  ...
# 1 * 9 = 9  2 * 9 = 18     ...  9 * 9 = 81

i = 1
while i < 10:
    j = 1
    while j < 10:
        print(f'{j} * {i} = {i*j}', end='\t')
        j += 1
    i += 1
    print()
```

```python
i = 1
while True:
    j = 1
    while True:
        print(f'{j} * {i} = {i*j}', end='\t')
        j += 1

        if j >= 10:
            break
    i += 1
    print()

    if i >= 10:
        break
```

```python
# 유저로부터 숫자를 입력받는다.
# 유저가 정상적인 숫자를 입력할 때까지 입력을 강제한다.
# 유저가 입력한 숫자에 *2 한 숫자를 출력한다.
# 단, 유저의 입력은 숫자로만 입력되지 않는다.

while True:
    user_input_value = input('숫자를 입력하세요: ')

    if user_input_value.isdecimal():
        break

user_input_value = int(user_input_value)
print(user_input_value*2)
```

```python
# 컴퓨터의 비밀번호는 486으로 설정되어있다고 한다.
# 컴퓨터의 비밀번호를 풀때까지 계속 입력을 시도하며
# 컴퓨터의 비밀번호를 맞추면 정답, 아니면 오답을 출력한다.
# 단, 사용자의 입력은 숫자가 아닐 수 있다.

while True:
    while True:
        user_input_value = input('숫자를 입력하세요: ')

        if user_input_value.isdecimal():
            break
        else:
            print('숫자로만 입력해주세요.')

    user_input_value = int(user_input_value)

    if user_input_value == 486:
        print('정답')
        break
    else:
        print('오답')
```

```python
# 정답은: 11
# 유저로부터 숫자를 입력받는다.
# 유저가 입력한 숫자와 정답을 비교하여 정답이 더 작은 숫자이면 down,
# 정답이 더 큰 숫자이면 up을 출력한다.
# 유저가 정답을 맞추면 정답이라는 멘트와 함께 프로그램을 종료한다.
# (optional) 유저가 몇 번에 맞췄나를 출력.
correct_number = 11
count = 0
while True:
    while True:
        user_input_value = input('숫자를 입력하세요: ')

        if user_input_value.isdecimal():
            break
        else:
            print('숫자로만 입력해주세요.')
    
    user_input_value = int(user_input_value)

    count += 1
    if user_input_value > correct_number:
        print('Down')
    elif user_input_value == correct_number:
        print('정답')
        print(f'정답까지 소요된 횟수: {count}')
        break
    else:
        print('Up')
```

###### for문과 문자열
for문과 문자열을 함께 사용하면 문자열 각 요소를  
순차적으로 하나씩 꺼내서 사용 가능.  

**사용법**  
for 문자 in 문자열:  
&nbsp;&nbsp;&nbsp;&nbsp;반복실행문  

```python
for char in 'hello world':
    print(char)
```

```python
# 모음만 출력

[char for char in 'hello world' if char in {'a', 'i', 'e', 'o', 'u'}]
```

```python
# 사용자로부터 비밀번호를 입력으면 그 비밀번호의 적합성을 구현하는 프로그램을 구현하려고 한다.
# 비밀번호는 최소 8자이며 반드시 1개 이상의 영문과 숫자의 조합으로 이루어져야 한다.
# 위의 코드를 구현하라.

user_input_value = input('비밀번호를 입력하세요: ')

if len(user_input_value) >= 8:
    char_count = 0
    num_count = 0
    for char in user_input_value:
        if char.isdecimal():
            num_count +=1
        elif char.isalpha():
            char_count +=1
    
    if (num_count>=1) and (char_count>=1):
        print('사용할 수 있는 비밀번호입니다.')
    else:
        print('숫자와 영문 조합을 사용해야 합니다.')
else:
    print('8자 미만의 비밀번호는 사용할 수 없습니다.')
```

## 5.3 기타 제어문

### 5.3.1 break

break문은 while문이나 for문과 같은 반복문을
강제로 종료하고자 할 때 사용하는 제어문

반복문 내에서 break문이 나타나면 곧바로 break문이 포함된 반복문은 종료  

```python
total = 0
for i in range(1, 11):
    if i%5 == 0:
        break

    total += i

print(total)
```

### 5.3.2 continue

continue문은 반복문의 시작 지점으로 제어의 흐름을 옮기는 역할을 수행.  

만약 1에서 100부터의 합을 구한다고 할 때 3의 배수를  
합계에서 제외하고 싶은 경우 등에 사용

```python
total = 0
for i in range(1, 11):
    if i%5 == 0:
        continue

    total += i

print(total)
```

### 5.3.3 return

return은 함수 끝에 입력돼서 해당 함수를 호출한 곳으로   
데이터를 반환하는 역할 또는, 함수의 종료를 알리는 기능.

```python
def test():
    total = 0
    for i in range(1, 11):
        if i%5 == 0:
            return

        total += i
        print(total)

    print(total)
```

```python
test()
```

```python
# 사용자로부터 임의의 양의 정수를 하나 입력받은 뒤
# 그 숫자 만큼 과일 이름을 입력받아 basket 리스트에 저장하는 프로그램 구현
# 단, basket 내에는 중복된 값이 입력될 수 없다.

max_basket_len = int(input('basket 길이를 입력하세요: '))
basket = []

for _ in range(max_basket_len):
    fruit_name = input('과일을 입력해주세요: ')

    while fruit_name in basket:
        fruit_name = input('중복되지 않은 과일을 입력해주세요: ')

    basket.append(fruit_name)
```

## 5. Codeup 1295. 
reference: [ASCII](https://www.vlsifacts.com/wp-content/uploads/2023/02/ASCII-Code.png) <br>
ord 함수 활용

```python
user_input = input()

# A ~ Z: 65 ~ 90
# a ~ z: 97 ~ 122

for char in user_input:
    if 65 <= ord(char) <= 90:
        print(char.lower(), sep='', end='')
    elif 97 <= ord(char) <= 122:
        print(char.upper(), sep='', end='')
    else:
        print(char, sep='', end='')
```
